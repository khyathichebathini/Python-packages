```python
import matplotlib.pyplot as plt
# data
x = [1,2,3,4,5,6,7,8]
y1 = [1,2,3,4,5,6,7,8]
y2 = [1,2,3,4,5,6,7,8]
# area chart
plt.fill_between(x, y1, y2, color='blue', alpha=0.4 , label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o') 
plt.plot(x, y2, label='Line 2', marker='o') 
#label and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('area chart')
# legend and display
plt.legend()
plt.show()
```
